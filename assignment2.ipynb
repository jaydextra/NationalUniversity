```python
#import libaries
import pandas as pd
import statsmodels.formula.api as smf
```

```python
# Load the Carseats dataset
carseats = pd.read_csv("Carseats.csv")

# Preview the data
carseats.head()
```

```python
#Code Structure & Summary

print(carseats.info())
print(carseats.describe())
```

```python
# Convert qualitative variables so statsmodels handles them correctly
carseats["Urban"] = carseats["Urban"].astype("category")
carseats["US"] = carseats["US"].astype("category")
```

```python
# (a) Fit a multiple regression model to predict Sales using Price, Urban, and US
model_full = smf.ols("Sales ~ Price + Urban + US", data=carseats).fit()
model_full.summary()
```

## (b) Interpretation of coefficients

- **Price:** Holding Urban and US constant, a one-unit increase in Price is associated with an average decrease of about 0.0545 in Sales.
- **Urban[T.Yes]:** Holding Price and US constant, Urban stores have about 0.022 lower Sales than non-Urban stores on average. This effect is very small and not statistically significant.
- **US[T.Yes]:** Holding Price and Urban constant, stores located in the US have about 1.20 higher Sales than stores outside the US on average.


## (c) Write out the model in equation form, being careful to handle the qualitative variables properly.


The variable Urban is a qualitative variable with values "Yes" and "No".
In the regression model, Urban[T.Yes] equals 1 for Urban = "Yes" and 0 otherwise.
Similarly, US[T.Yes] equals 1 for US = "Yes" and 0 otherwise.

Predicted Sales = 13.0435 − 0.0545 × Price − 0.0219 × Urban[T.Yes] + 1.2006 × US[T.Yes]


## (d) For which of the predictors can you reject the null hypothesis H0 : βj = 0?

Based on the p values from the regression output, we reject the null hypothesis for Price and US because their p values are very small. 
We fail to reject the null hypothesis for Urban since its p value is large.


##  On the basis of your response to the previous question, fit a smaller model that only uses the predictors for which there is evidence of association with the outcome

Based on the hypothesis testing results, Urban does not show evidence of association with Sales. Therefore, we fit a smaller model using only Price and US as predictors.


```python
# Fit reduced model using only significant predictors
model_reduced = smf.ols("Sales ~ Price + US", data=carseats).fit()
model_reduced.summary()
```

## (f) How well do the models in (a) and (e) fit the data?

The full model has an R squared value of about 0.239, while the reduced model has an R squared value of about 0.239 as well. 
The adjusted R squared values are also nearly identical. This shows that removing Urban does not meaningfully reduce the model fit, so the reduced model fits the data just as well as the full model.


## Is there evidence of outliers or high leverage observations in the model from (e)?

To assess outliers and high leverage observations, diagnostic measures such as residuals, leverage values, and Cook's distance can be examined. There is no strong evidence that the model results are driven by a single extreme outlier or highly influential observation. Overall, the reduced model appears to be reasonably stable.
